```python
import pandas as pd
```

```python
ACC = pd.read_csv("ACC 4.csv")
```

```python
ACC
```

```python
GYR = pd.read_csv("GYR 4.csv")
```

```python
GYR
```

```python
temptime = []
for i in range(len(ACC)):
    temptime.append(int(ACC["time"][i]/100))
```

```python
ACC['UpdatedTime'] = temptime
```

```python
ACC = ACC.drop(columns=["time"])
```

```python
temptime = []
for i in range(len(GYR)):
    temptime.append(int(GYR["time"][i]/100))
```

```python
GYR['UpdatedTime'] = temptime
```

```python
GYR = GYR.drop(columns=["time"])
```

```python
ACC
```

```python
GYR
```

```python
Maindf = pd.merge(ACC, GYR, on='UpdatedTime')
```

```python
Maindf
```

```python
Label = []
for i in range(0,len(Maindf),1):
    if i<1400:
        Label.append("Jog")
    else:
        Label.append("walk")
```

```python
Maindf["Label"] = Label
```

```python
Maindf.iloc[0:10,:]
```

```python
acc_x = []
acc_y = []
acc_z = []
gyr_x = []
gyr_y = []
gyr_z = []
time = []
Label = []
for i in range(len(Maindf)-1):
    if(Maindf.UpdatedTime[i] == Maindf.UpdatedTime[i+1]):
        acc_x.append(Maindf.x_x[i])
        acc_y.append(Maindf.y_x[i])
        acc_z.append(Maindf.z_x[i])
        gyr_x.append((Maindf.x_y[i]+Maindf.x_y[i+1])/2)
        gyr_y.append((Maindf.y_y[i]+Maindf.y_y[i+1])/2)
        gyr_z.append((Maindf.x_y[i]+Maindf.x_y[i+1])/2)
        time.append(Maindf.UpdatedTime[i])
        Label.append(Maindf.Label[i])
```

```python
len(acc_x)
```

```python
df = pd.DataFrame(acc_x,columns=["acc_x"])
df["acc_y"] = acc_y
df["acc_z"] = acc_z
df["gyr_x"] = gyr_x
df["gyr_y"] = gyr_y
df["gyr_z"] = gyr_z
df["time"] = time
df["Label"] = Label
```

```python
df
```

```python
acc_x1 = []
acc_y1 = []
acc_z1 = []
gyr_x1 = []
gyr_y1 = []
gyr_z1 = []
time1 = []
Label1= []
for i in range(len(df)-1):
    if(df.time[i] == df.time[i+1]):
        acc_x1.append(df.acc_x[i])
        acc_y1.append(df.acc_y[i])
        acc_z1.append(df.acc_z[i])
        gyr_x1.append((df.gyr_x[i]+df.gyr_x[i+1])/2)
        gyr_y1.append((df.gyr_y[i]+df.gyr_y[i+1])/2)
        gyr_z1.append((df.gyr_z[i]+df.gyr_z[i+1])/2)
        time1.append(df.time[i])
        Label1.append(df.Label[i])
    else:
        acc_x1.append(df.acc_x[i])
        acc_y1.append(df.acc_y[i])
        acc_z1.append(df.acc_z[i])
        gyr_x1.append(df.gyr_x[i])
        gyr_y1.append(df.gyr_y[i])
        gyr_z1.append(df.gyr_z[i])
        time1.append(df.time[i])
        Label1.append(df.Label[i])
```

```python
len(acc_x1)
```

```python
df = pd.DataFrame(acc_x1,columns=["acc_x1"])
df["acc_y1"] = acc_y1
df["acc_z1"] = acc_z1
df["gyr_x1"] = gyr_x1
df["gyr_y1"] = gyr_y1
df["gyr_z1"] = gyr_z1
df["time1"] = time1
df["Label1"] = Label1
```

```python
df
```

```python
for i in range(len(df)-1):
    if(df.time1[i] == df.time1[i+1]):
        df = df.drop(labels=[i])
```
